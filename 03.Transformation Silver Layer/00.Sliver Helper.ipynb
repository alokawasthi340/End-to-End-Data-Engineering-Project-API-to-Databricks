```python
from pyspark.sql.types import StringType
from pyspark.sql.functions import udf
from datetime import timedelta, datetime
import re

def parse_duration(duration_str):
    if not duration_str or duration_str == "":
        return "00:00:00"
    duration_str = duration_str.replace("P", "").replace("T", "")
    values = {"D": 0, "H": 0, "M": 0, "S": 0}
    pattern = r'(?:(\d+)D)?(?:(\d+)H)?(?:(\d+)M)?(?:(\d+)S)?'
    match = re.match(pattern, duration_str)
    if match:
        d, h, m, s = match.groups()
        values["D"] = int(d) if d else 0
        values["H"] = int(h) if h else 0
        values["M"] = int(m) if m else 0
        values["S"] = int(s) if s else 0
    total_duration = timedelta(
        days=values["D"], hours=values["H"], minutes=values["M"], seconds=values["S"]
    )
    total_duration = (datetime.min + total_duration).time()
    return str(total_duration)
```
