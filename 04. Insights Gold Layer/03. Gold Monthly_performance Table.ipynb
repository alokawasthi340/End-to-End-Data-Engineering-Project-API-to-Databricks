# Gold Table: gold.monthly_performance

## Objective

Create a monthly aggregated table to analyze publishing activity and video performance over time.

## Source
- gold.video_performance

## Grain
- One row per Year and Month

## Group By

- publish_year
- publish_month

## Columns

| Column | Description |
|---------|-------------|
| publish_year | Publication year |
| publish_month | Publication month |
| total_videos | Number of published videos |
| total_views | Sum of views |
| total_likes | Sum of likes |
| total_comments | Sum of comments |
| avg_views | Average views |
| avg_likes | Average likes |
| avg_comments | Average comments |

## Business Purpose

- Track monthly publishing trends.
- Analyze monthly audience growth.
- Compare performance across months and years.

```python
from pyspark.sql import functions as f
```

```python
source_table="youtube.gold.video_performance"
target_table="youtube.gold.monthly_performance"
```

```python
monthly_performance_df = (
    spark.read.table(source_table)
    .groupBy("publish_year", "publish_month")
    .agg(
        f.count("video_id").alias("total_videos"),
        f.sum("views").alias("total_views"),
        f.sum("likes").alias("total_likes"),
        f.sum("comments").alias("total_comments"),
        f.round(f.avg("views"),2).alias("avg_views"),
        f.round(f.avg("likes"),2).alias("avg_likes"),
        f.round(f.avg("comments"),2).alias("avg_comments"),
    )
)
```

```python
(
    monthly_performance_df.write.mode("overwrite").saveAsTable(target_table)
)
```

```python
%sql
select * from youtube.gold.monthly_performance limit 10;
```
